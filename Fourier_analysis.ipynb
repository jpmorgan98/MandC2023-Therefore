# Fourier Analysis of:
Time Dependent Multiple Balance - Simple Corner Balance discrtization scheme of the transient monoenergetic $S_N$ Neutron Transport Equation in slab geometry.

## Derrivation:
To ensure that our combination of higher order discretization schemes is still unconditionally stable, we performed a Fourier analysis to find the slowest-converging mode of the method.
Assuming no scattering, no sources, and making physical assumptions (only constant positive values of $\Delta x$, $\Delta t$, $\Sigma$, and $v$) we start by defining our Fourier ansätze as:
\begin{align}
    \psi_{k+1/2,j,L} &= \theta^{k+1}a e^{i\omega j}
    &
    \psi_{k+1/2,j,R} &= \theta^{k+1}b e^{i\omega j}
\end{align}

\begin{align}
    \psi_{k,j,L} &= \theta^{k}c e^{i\omega j}
    &
    \psi_{k,j,R} &= \theta^{k}d e^{i\omega j}
\end{align}

\begin{align}
    \psi_{k-1/2,j,L} &= \theta^{k}a e^{i\omega j}
    &
    \psi_{k-1/2,j,R} &= \theta^{k}b e^{i\omega j}
\end{align}
where $k$ is the mode number, $i$ is the imaginary number, $\omega$ is the frequency of the mode, $\theta$ is the , and $j$ is the specific cell.
Substituting our ansätze into the discrtization scheme and simplify to form:
\begin{equation}
    \frac{\Delta}{x} \frac{1}{v \Delta t} (\theta a - a) + \mu \left(\frac{c+d}{2} - d e ^{-i\omega} \right) + \frac{\Sigma\Delta x }{2}c = 0
\end{equation}

\begin{equation}
    \frac{\Delta}{x} \frac{1}{v \Delta t} (\theta b - b) + \mu \left( ce^{i\omega} - \frac{c+d}{2} \right) + \frac{\Sigma\Delta x }{2}d = 0
\end{equation}

\begin{equation}
    \frac{\Delta}{x} \frac{1}{v \Delta t} (\theta a - c) + \theta\mu \left( \frac{a+b}{2} - b e^{-i\omega} \right) + \frac{\Sigma\Delta x }{2}a \theta = 0
\end{equation}

\begin{equation}
    \frac{\Delta}{x} \frac{1}{v \Delta t} (\theta b - d) + \mu \left( d e^{i\omega} + \frac{c+d}{2} \right) + \frac{\Sigma\Delta x}{2} a\theta = 0 \;,
\end{equation}
combining Eq.~\eqref{eq:eig1} into \eqref{eq:eig2}:
\begin{equation} 
    \begin{bmatrix}
    \frac{\mu}{2}+\frac{\Sigma \Delta x}{2} & \mu (\frac{1}{2} - e^{-i\omega}) \\
    \mu (e^{i\omega} - \frac{1}{2}) & - \frac{\mu}{2}+\frac{\Sigma \Delta x}{2}
    \end{bmatrix}
    \begin{bmatrix}
        c \\ d
    \end{bmatrix}
    =
    \begin{bmatrix}
        \frac{\Delta x}{2}\frac{1}{x\Delta t} & 0 \\
        0 & \frac{\Delta x}{2}\frac{1}{x\Delta t}
    \end{bmatrix}
    (1-\theta)
    \begin{bmatrix}
        a \\ b
    \end{bmatrix}
\end{equation}

\begin{equation}
    \begin{bmatrix}
        c\\d
    \end{bmatrix}
    = \boldsymbol{\mathcal{A}}^{-1} \frac{\Delta x}{2}\frac{1}{v\Delta t}(1-\theta)
    \begin{bmatrix}
        a\\b
    \end{bmatrix} \;,
\end{equation}
where
\begin{equation*}
    \boldsymbol{\mathcal{A}} = 
    \begin{bmatrix}
    \frac{\mu}{2}+\frac{\Sigma \Delta x}{2} & \mu (\frac{1}{2} - e^{-i\omega}) \\
    \mu (e^{i\omega} - \frac{1}{2}) & - \frac{\mu}{2}+\frac{\Sigma \Delta x}{2}
    \end{bmatrix} \;.
\end{equation*}
Then, doing the same with Eq.~\eqref{eq:eig3} into \eqref{eq:eig4}:
\begin{equation}
    \begin{bmatrix}
    \frac{\Delta x}{2}\frac{2}{c\Delta t} + \frac{\mu}{2}+\frac{\Sigma \Delta x}{2} & \mu (\frac{1}{2} - e^{-i\omega}) \\
    \mu (e^{i\omega} - \frac{1}{2}) & \frac{\Delta x}{2}\frac{2}{c\Delta t} + \frac{\mu}{2}+\frac{\Sigma \Delta x}{2}
    \end{bmatrix}
    \theta 
    \begin{bmatrix}
        a \\ b
    \end{bmatrix}
    =
    \frac{\Delta x}{2}\frac{2}{v \Delta t}
    \begin{bmatrix}
        c \\ d
    \end{bmatrix}
\end{equation}

\begin{equation}
    \boldsymbol{\mathcal{B}}\theta
    \begin{bmatrix}
        a\\b
    \end{bmatrix}
    = \frac{\Delta x}{2} \frac{2}{v\Delta t} \left( \boldsymbol{\mathcal{A}}^{-1} \frac{\Delta x}{2} \frac{1}{v\Delta t} (1-\theta) \right) \;,
\end{equation}
where
\begin{equation*}
    \boldsymbol{\mathcal{B}} = 
    \begin{bmatrix}
        \frac{\Delta x}{2}\frac{2}{c\Delta t} + \frac{\mu}{2}+\frac{\Sigma \Delta x}{2} & \mu (\frac{1}{2} - e^{-i\omega}) \\
        \mu (e^{i\omega} - \frac{1}{2}) & \frac{\Delta x}{2}\frac{2}{c\Delta t} + \frac{\mu}{2}+\frac{\Sigma \Delta x}{2}
    \end{bmatrix} \;.
\end{equation*}
Substituting Eq.~\eqref{eq:f_a-2} into \eqref{eq:f_a-1} and rearranging:
\begin{equation}
    (\boldsymbol{\mathcal{B}}+\gamma \boldsymbol{\mathcal{A}}^{-1}) \theta
    \begin{bmatrix}
        a \\ b
    \end{bmatrix}
    = \frac{\Delta x}{2} \frac{2}{v \Delta t} \boldsymbol{\mathcal{A}}^{-1} \frac{\Delta x}{2} \frac{1}{v\Delta t}
\end{equation}
where $\gamma = \frac{\Delta x}{v\Delta t}  \frac{\Delta x}{2v\Delta t}$
This can then be more appropriately posed as an eigenfunction:
\begin{equation}
    \theta
    \begin{bmatrix}
        a \\ b
    \end{bmatrix}
    = (\boldsymbol{\mathcal{B}} + \gamma \boldsymbol{\mathcal{A}}^{-1}) (\gamma \boldsymbol{\mathcal{A}}^{-1})
    \begin{bmatrix}
        a \\b
    \end{bmatrix}
    = \Lambda
    \begin{bmatrix}
        a\\b
    \end{bmatrix} \;.
\end{equation}

When this system is solved for its eigenvalues ($\Lambda$) [numerically](https://github.com/CEMeNT-PSAAP/stability_analysis) it shows that a MB-SCB scheme is unconditionally stable over $\mu \in [-1, 1]$. 
While experimenting with this method we did find that, under some conditions, it can produce negative fluxes; however, the negative flux oscillations were critically damped and dissipated with time.





## Numerical Soultion
Below we present our numerical soultion to the MB-SCB fourier anayliss using the systems of equations represented above

```python
from ast import Interactive
from functools import cache
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors

sig_t = 1   # total cross section
mu = 1      # angle in space
v = 1       # velocity

# negitive egien function system (from derrivation)
def eig_val(dt, dx, omg, v, mu, sig_t):
    i = complex(0, 1)

    Acompp = np.array([[mu/2 + sig_t*dx/2,   mu*(.5-np.exp(-i*omg))],
                       [-mu/2,               mu/2 + sig_t*dx/2]])


    gamma = dx/(v*dt) * dx/(2*v*dt)

    I = np.array([[1,0],[0,1]])

    Acompp_sq = np.matmul(Acompp, Acompp) + dx/(v*dt)*Acompp + gamma*I

    eige_mat = gamma * np.linalg.inv(Acompp_sq)


    eig_val, stand_eig_mat = np.linalg.eig(eige_mat)

    return(np.max(np.abs(eig_val)))

# negitive egien function system (from derrivation)
def eig_neg(dt, dx, omg, v, mu, sig_t):
    i = complex(0, 1)

    Acompn = np.array([[-mu/2 + sig_t*dx/2,     mu/2],
                    [mu*(np.exp(i*omg)-.5), -mu/2 + sig_t*dx/2]])


    gamma = dx/(v*dt) * dx/(2*v*dt)

    I = np.array([[1,0],[0,1]])

    Acompn_sq = np.matmul(Acompn,  Acompn) + dx/(v*dt)*Acompn + gamma*I

    eige_mat = gamma * np.linalg.inv(Acompn_sq)

    eig_val, stand_eig_mat = np.linalg.eig(eige_mat)

    return(np.max(np.abs(eig_val)))

# compute the eigen values
def Acomp(DT, DX, omg, v, mu, sig_t, Neta, Ntau):
    Ap = np.zeros_like(DT)
    An = np.zeros_like(DT)
    for i in nb.prange(Neta):
        for j in nb.prange(Ntau):
            Ap[i,j] = eig_val(DT[i,j], DX[i,j], omg, v, mu, sig_t)
            An[i,j] = eig_neg(DT[i,j], DX[i,j], omg, v, -mu, sig_t)
    
    return(Ap, An)


Neta = 100 # how many values of eta (dt)
Ntau = 100 # how many values of tau (dx)
Nomg = 5   # on how many frequnce modes (omega) this will be the number of plots
dt  = np.linspace(0.001,5,Neta)
dx  = np.linspace(0.001,5,Ntau)
fac  = np.linspace(0.0,2.0,Nomg)[:-1]

DT, DX = np.meshgrid(dt, dx)

for k in range(Nomg-1):
    omg = fac[k]*np.pi
    print(k)

    # copmute the eigen values in all eta and tau
    Ap, An = Acomp(DT, DX, omg, v, mu, sig_t, Neta, Ntau)

    # if its greater then one no convergence
    An = An <= 1.0
    Ap = Ap <= 1.0

    # plotting in angle
    cmap = colors.ListedColormap(['crimson', 'dodgerblue'])
    plt.figure(figsize=(4,4))
    plt.pcolormesh(DT, DX, Ap, cmap=cmap)
    plt.title(r'$+\mu$ $\omega=%.2f\pi$'%fac[k])
    plt.ylabel(r'$\eta$')
    plt.xlabel(r'$\tau$')
    plt.grid()
    plt.gca().set_aspect('equal', 'box')
    plt.savefig('MB-SCBp_%i.png'%(k+1), bbox_inches = 'tight', dpi=1200, pad_inches = 0)
    cmap = colors.ListedColormap(['crimson', 'dodgerblue'])
    plt.figure(figsize=(4,4))
    plt.pcolormesh(DT, DX, An, cmap=cmap)
    plt.title(r'$-\mu$ $\omega=%.2f\pi$'%fac[k])
    plt.ylabel(r'$\eta$')
    plt.xlabel(r'$\tau$')
    plt.grid()
    plt.gca().set_aspect('equal', 'box')

# Red is stable!
```
