# Derrivation:
To ensure that our combination of higher order discretization schemes is still unconditionally stable, we performed a Fourier analysis to find the slowest-converging mode of the method.
Assuming no scattering, no sources, and making physical assumptions (only constant positive values of $\Delta x$, $\Delta t$, $\Sigma$, and $v$) we start by defining our Fourier ansätze as:
\begin{align}
    \psi_{k+1/2,j,L} &= \theta^{k+1}a e^{i\omega j}
    &
    \psi_{k+1/2,j,R} &= \theta^{k+1}b e^{i\omega j}
\end{align}

\begin{align}
    \psi_{k,j,L} &= \theta^{k}c e^{i\omega j}
    &
    \psi_{k,j,R} &= \theta^{k}d e^{i\omega j}
\end{align}

\begin{align}
    \psi_{k-1/2,j,L} &= \theta^{k}a e^{i\omega j}
    &
    \psi_{k-1/2,j,R} &= \theta^{k}b e^{i\omega j}
\end{align}
where $k$ is the mode number, $i$ is the imaginary number, $\omega$ is the frequency of the mode, $\theta$ is the , and $j$ is the specific cell.
Substituting our ansätze into Eqs.~\eqref{eq:scb-mb-a}, \eqref{eq:scb-mb-b}, \eqref{eq:scb-mb-c}, and \eqref{eq:scb-mb-d}, respectively, we simplify to form:
\begin{equation}
    \frac{\Delta}{x} \frac{1}{v \Delta t} (\theta a - a) + \mu \left(\frac{c+d}{2} - d e ^{-i\omega} \right) + \frac{\Sigma\Delta x }{2}c = 0
\end{equation}

\begin{equation}
    \frac{\Delta}{x} \frac{1}{v \Delta t} (\theta b - b) + \mu \left( ce^{i\omega} - \frac{c+d}{2} \right) + \frac{\Sigma\Delta x }{2}d = 0
\end{equation}

\begin{equation}
    \frac{\Delta}{x} \frac{1}{v \Delta t} (\theta a - c) + \theta\mu \left( \frac{a+b}{2} - b e^{-i\omega} \right) + \frac{\Sigma\Delta x }{2}a \theta = 0
\end{equation}

\begin{equation}
    \frac{\Delta}{x} \frac{1}{v \Delta t} (\theta b - d) + \mu \left( d e^{i\omega} + \frac{c+d}{2} \right) + \frac{\Sigma\Delta x}{2} a\theta = 0 \;,
\end{equation}
combining Eq.~\eqref{eq:eig1} into \eqref{eq:eig2}:
\begin{equation} 
    \begin{bmatrix}
    \frac{\mu}{2}+\frac{\Sigma \Delta x}{2} & \mu (\frac{1}{2} - e^{-i\omega}) \\
    \mu (e^{i\omega} - \frac{1}{2}) & - \frac{\mu}{2}+\frac{\Sigma \Delta x}{2}
    \end{bmatrix}
    \begin{bmatrix}
        c \\ d
    \end{bmatrix}
    =
    \begin{bmatrix}
        \frac{\Delta x}{2}\frac{1}{x\Delta t} & 0 \\
        0 & \frac{\Delta x}{2}\frac{1}{x\Delta t}
    \end{bmatrix}
    (1-\theta)
    \begin{bmatrix}
        a \\ b
    \end{bmatrix}
\end{equation}

\begin{equation}
    \begin{bmatrix}
        c\\d
    \end{bmatrix}
    = \boldsymbol{\mathcal{A}}^{-1} \frac{\Delta x}{2}\frac{1}{v\Delta t}(1-\theta)
    \begin{bmatrix}
        a\\b
    \end{bmatrix} \;,
\end{equation}
where
\begin{equation*}
    \boldsymbol{\mathcal{A}} = 
    \begin{bmatrix}
    \frac{\mu}{2}+\frac{\Sigma \Delta x}{2} & \mu (\frac{1}{2} - e^{-i\omega}) \\
    \mu (e^{i\omega} - \frac{1}{2}) & - \frac{\mu}{2}+\frac{\Sigma \Delta x}{2}
    \end{bmatrix} \;.
\end{equation*}
Then, doing the same with Eq.~\eqref{eq:eig3} into \eqref{eq:eig4}:
\begin{equation}
    \begin{bmatrix}
    \frac{\Delta x}{2}\frac{2}{c\Delta t} + \frac{\mu}{2}+\frac{\Sigma \Delta x}{2} & \mu (\frac{1}{2} - e^{-i\omega}) \\
    \mu (e^{i\omega} - \frac{1}{2}) & \frac{\Delta x}{2}\frac{2}{c\Delta t} + \frac{\mu}{2}+\frac{\Sigma \Delta x}{2}
    \end{bmatrix}
    \theta 
    \begin{bmatrix}
        a \\ b
    \end{bmatrix}
    =
    \frac{\Delta x}{2}\frac{2}{v \Delta t}
    \begin{bmatrix}
        c \\ d
    \end{bmatrix}
\end{equation}

\begin{equation}
    \boldsymbol{\mathcal{B}}\theta
    \begin{bmatrix}
        a\\b
    \end{bmatrix}
    = \frac{\Delta x}{2} \frac{2}{v\Delta t} \left( \boldsymbol{\mathcal{A}}^{-1} \frac{\Delta x}{2} \frac{1}{v\Delta t} (1-\theta) \right) \;,
\end{equation}
where
\begin{equation*}
    \boldsymbol{\mathcal{B}} = 
    \begin{bmatrix}
        \frac{\Delta x}{2}\frac{2}{c\Delta t} + \frac{\mu}{2}+\frac{\Sigma \Delta x}{2} & \mu (\frac{1}{2} - e^{-i\omega}) \\
        \mu (e^{i\omega} - \frac{1}{2}) & \frac{\Delta x}{2}\frac{2}{c\Delta t} + \frac{\mu}{2}+\frac{\Sigma \Delta x}{2}
    \end{bmatrix} \;.
\end{equation*}
Substituting Eq.~\eqref{eq:f_a-2} into \eqref{eq:f_a-1} and rearranging:
\begin{equation}
    (\boldsymbol{\mathcal{B}}+\gamma \boldsymbol{\mathcal{A}}^{-1}) \theta
    \begin{bmatrix}
        a \\ b
    \end{bmatrix}
    = \frac{\Delta x}{2} \frac{2}{v \Delta t} \boldsymbol{\mathcal{A}}^{-1} \frac{\Delta x}{2} \frac{1}{v\Delta t}
\end{equation}
where $\gamma = \frac{\Delta x}{v\Delta t}  \frac{\Delta x}{2v\Delta t}$
This can then be more appropriately posed as an eigenfunction:
\begin{equation}
    \theta
    \begin{bmatrix}
        a \\ b
    \end{bmatrix}
    = (\boldsymbol{\mathcal{B}} + \gamma \boldsymbol{\mathcal{A}}^{-1}) (\gamma \boldsymbol{\mathcal{A}}^{-1})
    \begin{bmatrix}
        a \\b
    \end{bmatrix}
    = \Lambda
    \begin{bmatrix}
        a\\b
    \end{bmatrix} \;.
\end{equation}

When this system is solved for its eigenvalues ($\Lambda$) numerically\footnote{\url{https://github.com/CEMeNT-PSAAP/stability_analysis}} it shows that a MB-SCB scheme is unconditionally stable over $\mu \in [-1, 1]$. 
While experimenting with this method we did find that, under some conditions, it can produce negative fluxes; however, the negative flux oscillations were critically damped and dissipated with time.
